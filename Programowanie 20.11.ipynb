## Zadanie 2

```python
x = 17
def dzielniki(x):
    lista_dzielnikow_1 = {i for i in range(1, int(x**0.5) + 1) if x % i == 0}
    return lista_dzielnikow_1|{x // i for i in lista_dzielnikow_1}

dzielniki(98)

slownik_dzielnikow = {x : dzielniki(x) for x in range(x)}
slownik_dzielnikow
```

## Zadanie 3 - odwracanie słownika

```python
dict = {1:'a', 2:'c', 3:'f', 7:'g'}

def odwroc_slownik(slownik):
    odp = {}
    for klucz in slownik:
        try:
            hash(slownik[klucz])
        except TypeError:
            print('Nie da się odwrócić')
            return None
        odp[slownik[klucz]] = klucz
    return odp

odwroc_slownik(dict)
```
