```python
x_1 = np.linspace(0, np.pi, 100)
```

```python
np.sin(x_1)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_1, np.sin(x_1), color = 'black', ls = '-.', label = '$\sin{x}$')
plt.plot(x_1, np.cos(x_1), color = 'red', ls = '--', label = '$\cos{x}$')
plt.grid()
plt.show()
```

```python
x_plt = np.linspace(0, 2 * np.pi, 100)
plt.plot(x_plt, np.sin(x_plt), color = 'red', label = '$\sin({x})$')
plt.plot(x_plt, np.cos(x_plt), color = 'blue', label = '$\cos({x})$')
plt.plot(x_plt, np.zeros(len(x_plt)), color = 'grey', alpha=0.5)

plt.legend(loc = 'upper right')
plt.xticks([0, np.pi, 2*np.pi], ['0', '$\pi$', '$2\pi$'])
plt.ylim((-1.5, 2))
plt.tick_params(axis='x', direction='in')
plt.tick_params(axis='y', direction='in')
plt.tick_params(labeltop = True)
#plt.label('x')
#plt.title('Wykres')
plt.yticks([-1, 0, 1])
plt.margins(x=0, y=0)
#help(plt.margins)
plt.fill_between(x_plt, np.sin(x_plt), np.cos(x_plt), color = 'blue', alpha=0.25)

plt.show()
```

```python
plt.subplot(221)    #   (wiersze, kolumny, indeks)
plt.plot(x_1, np.sin(x_1))

plt.subplot(222)
plt.plot(x_1, np.cos(x_1))

plt.subplot(223)
plt.plot(x_1, np.sin(x_1) - np.cos(x_1))

plt.subplot(224)
plt.plot(x_1, np.sin(5*x_1))
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 1000)
r = 2*(1 + np.cos(t))

ax = plt.subplot(111, projection = 'polar')
ax.plot(t, r)
plt.ylim((0, 10))
#ax.set_rlim(0, 10)
plt.show()
```

```python
X, Y = np.meshgrid(np.linspace(-2, 2, 1000), np.linspace(-2, 2, 1000))
Z = np.exp(-(X**2 + Y**2))

ax = plt.subplot(111, projection = '3d')
ax.plot_surface(X,Y,Z)
plt.show()
```
